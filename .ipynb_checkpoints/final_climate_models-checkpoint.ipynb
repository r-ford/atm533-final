# 1. Visualizing sea-surface temperature in a 1pctCO2 experiment

```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
from cartopy import crs as ccrs, feature as cfeature
import datetime as dt
```

### In this section, we will visualize the change in SST for a climate model given a 1% per year increase in CO$_2$ until a doubling of CO$_2$ is reached (1pctCO2).

We will use NCAR's Community Earth System Model Version 2 (CESM2), which participates in the Coupled Model Intercomparison Project Phase 6 (CMIP6). Outputs from various experiments can be downloaded remotely. We will download the SST data (here named `tos`) for the 1pctCO2 experiment, as well as for the pre-industrial control run (piControl) for comparison. 

Another common experiment is an abrupt quadrupling of CO$_2$ (4xCO2), which makes the model equilibrate faster so that the equilibrium state of the climate can be examined. A 1pctCO2 experiment is more realistic and allows us to analyze the transient warming.

```python
filepath = 'https://esgf-data.ucar.edu/thredds/dodsC/esg_dataroot/CMIP6/CMIP/NCAR/CESM2/'
exp_path = '1pctCO2/r1i1p1f1/Omon/tos/gr/v20190425/tos_Omon_CESM2_1pctCO2_r1i1p1f1_gr_000101-005012.nc'
ctl_path = 'piControl/r1i1p1f1/Omon/tos/gr/v20190320/tos_Omon_CESM2_piControl_r1i1p1f1_gr_000101-009912.nc'

piControl = xr.open_dataset(filepath + ctl_path)
pctCO2 = xr.open_dataset(filepath + exp_path)
```

In these datasets, the SST is given at every point (of the ocean) on a 1x1 lat-lon grid once a month. As the `time` coordinate shows, the piControl dataset spans 99 years (Jan 15, 0001 to Dec 15, 0099) and the 1pctCO2 dataset spans 50 (Jan 15, 0001 to Dec 15, 0050).

```python
piControl
```

```python
pctCO2
```

First, let's take a quick look at the data. Using Xarray's `.plot()` method is a convenient way to check the data before making the plot look nicer with Matplotlib and Cartopy.

```python
piControl.isel(time=0).tos.plot()
```

Now let's look at the SST around the end of year 50 (specifically Dec 15, 0050) for both the control and experimental runs.

```python
date = dt.date(50, 12, 15)
datestr = date.strftime('00%Y-%m-%d')
piControl_SST_50y = piControl.sel(time=datestr).tos.isel(time=0)
pctCO2_SST_50y = pctCO2.sel(time=datestr).tos.isel(time=0)
```

Getting the lats and lons for convenience&#151;they are the same for both datasets:

```python
lats = piControl.lat
lons = piControl.lon
```

```python
fig1 = plt.figure(1, dpi=130, figsize=(12, 5))

proj = ccrs.Mollweide(central_longitude=180)
PC = ccrs.PlateCarree()

ax1 = plt.subplot(2, 1, 1, projection=proj)
ctl_plot = ax1.contourf(lons, lats, piControl_SST_50y, transform=PC, cmap='seismic', vmin=-35, vmax=35, levels=20)
ax1.coastlines()
ax1.set_title('piControl')

ax2 = plt.subplot(2, 1, 2, projection=proj)
exp_plot = ax2.contourf(lons, lats, pctCO2_SST_50y, transform=PC, cmap='seismic', vmin=-35, vmax=35, levels=20)
ax2.coastlines()
ax2.set_title('1pctCO2')

fig1.colorbar(ctl_plot, shrink=0.8, ax=[ax1, ax2], label='SST (deg C)')
fig1.suptitle('$t=$'+datestr)
```

Of course, this is only for December. Let's now plot the yearly average for year 50. We will use Xarray's `.mean()` method for this.

```python
piControl_SST_50y_annual = piControl.sel(time=slice('0050-01-15', '0050-12-15')).mean(dim='time').tos
pctCO2_SST_50y_annual = pctCO2.sel(time=slice('0050-01-15', '0050-12-15')).mean(dim='time').tos
```

Recreating the plot above with the yearly average:

```python
fig2 = plt.figure(2, dpi=130, figsize=(12, 5))

ax3 = plt.subplot(2, 1, 1, projection=proj)
ctl_plot = ax3.contourf(lons, lats, piControl_SST_50y_annual, transform=PC, cmap='seismic', vmin=-35, vmax=35, levels=20)
ax3.coastlines()
ax3.set_title('piControl')

ax4 = plt.subplot(2, 1, 2, projection=proj)
exp_plot = ax4.contourf(lons, lats, pctCO2_SST_50y_annual, transform=PC, cmap='seismic', vmin=-35, vmax=35, levels=20)
ax4.coastlines()
ax4.set_title('1pctCO2')

fig2.colorbar(ctl_plot, shrink=0.8, ax=[ax3, ax4], label='SST (deg C)')
fig2.suptitle('$t=00$'+str(date.year))
```

Latitude seems to have the greatest effect on SST, so let's look at the zonal mean of year 50 by taking the mean over longitude for the yearly average we already calculated.

Note that over land, the datasets use the fill value `nan`. This could cause a problem if we take a spatial average and include this area, since it would bring down our average below its actual value. According to the [Xarray documentation](http://xarray.pydata.org/en/stable/generated/xarray.DataArray.mean.html), `.mean()` will by default skip missing values if our data consists of floats, which it does.

```python
piControl_SST_50y_annual_zonal = piControl_SST_50y_annual.mean(dim='lon')
pctCO2_SST_50y_annual_zonal = pctCO2_SST_50y_annual.mean(dim='lon')
```

```python
fig3 = plt.figure(3, dpi=130, figsize=(6, 3))

axz = plt.subplot(1, 1, 1)
axz.plot(lats, piControl_SST_50y_annual_zonal, color='steelblue', label='piControl')
axz.plot(lats, pctCO2_SST_50y_annual_zonal, color='seagreen', label='1pctCO2')
axz.set_xlim(-77.5, 89.5)
axz.set_xlabel('Latitude')
axz.set_ylabel('Temperature (deg C)')
axz.set_title('Zonal average SST')
axz.grid()
plt.legend()
```

Finally, let's plot the timeseries for the global average SST to see the trend. We also need to weight each point by the cosine of the latitude, since the area of a 1x1 lat-lon gridcell decreases with latitude as such. Xarray has a `.weighted()` method that will do this for us, provided we give it the weights.

```python
def global_average(data):
    weights = np.cos(np.deg2rad(data.lat))
    data_weighted = data.weighted(weights)
    return data_weighted.mean(dim=['lat', 'lon'])
```

Here we are taking the first 50 years of the control run to match the duration of the experimental run. An annual rolling mean is also created using Xarray's `.rolling()` method.

```python
piControl_first50 = piControl.sel(time=slice('0001-01-15', '0050-12-15'))
times = piControl_first50.time

piControl_SST_ga = global_average(piControl_first50.tos)
pctCO2_SST_ga = global_average(pctCO2.tos)

piControl_SST_ga_rolling = piControl_SST_ga.rolling(time=12, center=True).mean()
pctCO2_SST_ga_rolling = pctCO2_SST_ga.rolling(time=12, center=True).mean()
```

```python
fig4 = plt.figure(4, dpi=130, figsize=(10, 5))

ax5 = plt.subplot(1, 1, 1)
ax5.plot(times, piControl_SST_ga, '--', color='steelblue', linewidth=1)
ax5.plot(times, pctCO2_SST_ga, '--', color='seagreen', linewidth=1)
ax5.plot(times, piControl_SST_ga_rolling, color='steelblue', linewidth=2, label='piControl')
ax5.plot(times, pctCO2_SST_ga_rolling, color='seagreen', linewidth=2, label='1pctCO2')
ax5.set_xlim(times[0].values, times[-1].values)
ax5.set_yticks([17.5, 18.0, 18.5, 19.0, 19.5])
ax5.set_xlabel('Year')
ax5.set_ylabel('Temperature (deg C)')
ax5.set_title('Global average SST')
plt.legend()
plt.grid()
```

The trend is clearer now, and the annual rolling mean allows us to filter out the seasonal variations.

Just to demonstrate, this can also be plotted with Pandas by converting the Xarray Dataset to a Pandas DataFrame.

```python
piControl_pd = piControl_SST_ga.to_dataframe()
pctCO2_pd = pctCO2_SST_ga.to_dataframe()
```

```python
plt.plot(piControl_pd.index, piControl_pd.tos, label='piControl')
plt.plot(pctCO2_pd.index, pctCO2_pd.tos, label='1pctCO2')
plt.xlabel('Year')
plt.ylabel('Temperature (deg C)')
plt.title('Global average SST')
plt.legend()
```

# 2. Visualizing near-surface air temperature in a 4xCO2 experiment

### In this section, we will compare the near-surface air temperature between three similar CMIP5 models that receive an abrupt quadrupling of CO$_2$ (4xCO2).

The models used here are variations of the Max Planck Institute for Meteorology Earth System Model. MPI-ESM-LR (low resolution) is the base model, MPI-ESM-MR (mixed resolution) has more atmospheric pressure levels and a finer ocean grid, and MPI-ESM-P (paleo) has prescribed vegetation instead of dynamic ([details here](https://agupubs.onlinelibrary.wiley.com/doi/full/10.1002/jame.20038)).

The variable name for near-surface (2 m) air temperature is `tas`.

```python
ESM_LR = xr.open_dataset('https://esgf1.dkrz.de/thredds/dodsC/cmip5/cmip5/output1/MPI-M/MPI-ESM-LR/abrupt4xCO2/mon/atmos/Amon/r1i1p1/v20120602/tas/tas_Amon_MPI-ESM-LR_abrupt4xCO2_r1i1p1_185001-199912.nc')
ESM_MR = xr.open_dataset('https://esgf1.dkrz.de/thredds/dodsC/cmip5/cmip5/output1/MPI-M/MPI-ESM-MR/abrupt4xCO2/mon/atmos/Amon/r1i1p1/v20120602/tas/tas_Amon_MPI-ESM-MR_abrupt4xCO2_r1i1p1_185001-199912.nc')
ESM_P = xr.open_dataset('https://esgf1.dkrz.de/thredds/dodsC/cmip5/cmip5/output1/MPI-M/MPI-ESM-P/abrupt4xCO2/mon/atmos/Amon/r1i1p1/v20120602/tas/tas_Amon_MPI-ESM-P_abrupt4xCO2_r1i1p1_185001-199912.nc')
```

Let's use the global mean function from before, convert them to Pandas DataFrames, and merge them into one DataFrame.

```python
LR_TAS_ga = global_average(ESM_LR.tas).to_dataframe()
MR_TAS_ga = global_average(ESM_MR.tas).to_dataframe()
P_TAS_ga = global_average(ESM_P.tas).to_dataframe()
```

```python
TAS_ga_temp = pd.merge(LR_TAS_ga.tas, MR_TAS_ga.tas, on='time')
TAS_ga = pd.merge(TAS_ga_temp, P_TAS_ga.tas, on='time')
TAS_ga.columns = ['LR_tas', 'MR_tas', 'P_tas']
```

Here we plot the annual rolling mean TAS for the three models, as well as their average.

```python
fig5 = plt.figure(5, dpi=130, figsize=(10, 5))
plt.plot(TAS_ga.rolling(window=12).mean(), '--', linewidth=1, label=TAS_ga.columns)
plt.plot(TAS_ga.rolling(window=12).mean().mean(axis=1), color='k', linewidth=2, label='Mean_tas')
plt.xlim(TAS_ga.index[12], TAS_ga.index[-1])
plt.xlabel('Year')
plt.ylabel('Temperature (K)')
plt.title('2 m air temperature after 4xCO2')
plt.legend()
plt.grid()
```
